```python
df.drop_duplicates(subset=['review'], keep='first', inplace=True)
df.reset_index(drop=True, inplace=True)
df.head(20)
```

```python
df["review"] = df["review"].str.lower()
df["review"] = df["review"].str.replace(r"<.*?>", "", regex=True)
```

```python
df.head()
```

```python
X=df["review"]
y=df["sentiment"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=61)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(lowercase=True, stop_words='english',ngram_range=(1,2))
X_train=vectorizer.fit_transform(X_train)
X_test=vectorizer.transform(X_test)
```

```python
vectorizer.vocabulary_
```

```python
vectorizer.get_feature_names_out()
```

```python
y.unique()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.fit_transform(y_test)
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(solver='saga',max_iter=1000,penalty='l1')
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred, target_names=['negative', 'positive'])
print(report)
```
